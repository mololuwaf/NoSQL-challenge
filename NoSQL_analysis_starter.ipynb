### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = query ={'scores.Hygiene': 20}
results = establishments.find(query)  

# Use count_documents to display the number of documents in the result
print(establishments.count_documents(query))
# Display the first document in the results using pprint
hygiene_20 = list(results)
pprint(hygiene_20)
```

```python
# Convert the result to a Pandas DataFrame
hygiene_20_df = pd.DataFrame(establishments.find(query))
# Display the number of rows in the DataFrame
print(len(hygiene_20_df))
# Display the first 10 rows of the DataFrame
hygiene_20_df.head()
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query_2 = {'$and':[{'LocalAuthorityName': {'$regex':'London'}},
                  {'RatingValue':{'$gte':'4'}}
                  ]}


# Use count_documents to display the number of documents in the result
print(establishments.count_documents(query_2))
# Display the first document in the results using pprint
pprint(list(establishments.find(query_2)))
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(establishments.find(query_2))
# Display the number of rows in the DataFrame
print(len(df))
# Display the first 10 rows of the DataFrame
df.head()
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.49014200
longitude = 0.08384000

#parameters
latitude_1 = establishments.find_one({'BusinessName':'Penang Flavours'})['geocode']['latitude'] - degree_search
latitude_2 = establishments.find_one({'BusinessName':'Penang Flavours'})['geocode']['latitude'] + degree_search
longitude_1 = establishments.find_one({'BusinessName':'Penang Flavours'})['geocode']['longitude'] - degree_search
longitude_2 = establishments.find_one({'BusinessName':'Penang Flavours'})['geocode']['longitude'] + degree_search


query = {
    "geocode.longitude": {"$gte": longitude_1, "$lte": longitude_2},
    "geocode.latitude": {"$gte": latitude_1, "$lte": latitude_2},
    "RatingValue": "5"
}

output = establishments.find(query).sort("scores.Hygiene", -1)

for i in output:
    
    output_1 = list(output)

pprint(output_1)
```

```python
# Convert result to Pandas DataFrame
df_2 = pd.DataFrame(output_1)
df_2.head()
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
match_2 = {'$match':{'scores.Hygiene':{'$eq': 0}}}
group_2 = {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}}
sort_2 = {"$sort":{'count': -1} } 
pipeline = [match_2,group_2,sort_2]
# Print the number of documents in the result
results_3 = list(establishments.aggregate(pipeline))
print(len(results_3))
# Print the first 10 results
pprint(results_3[0:10])
```

```python
# Convert the result to a Pandas DataFrame
zero_hygiene = pd.DataFrame(results_3)
# Display the number of rows in the DataFrame
print(f'Number of rows: {len(zero_hygiene)}')
# Display the first 10 rows of the DataFrame
zero_hygiene.head(10)
```
